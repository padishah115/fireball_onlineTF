```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import json
```

```python
# idk how long the master csv is going to get - at some point might be worth writing a function to only load new lines added to the end of the master csv
#  and append these to the DataFrame to save loading the entire file after every shot

# load master csv into dataframe
folder = "example_data"
filename = "HRMT64_timestamps_shots_camerafiles.csv"

#Create dataframe from camera timestamps .csv
df = pd.read_csv(f"./{folder}/{filename}")

#Rename (unnamed) first column of df to "TIMESTAMP"
df = df.rename(columns={df.keys()[0]: "Timestamp"})

# Search for rows with no shot number, remove them from our list.
# Reset indices for the non-empty shots, then remove the annoying "index" column.
shots_df = df[df["Shot number"].notna()].reset_index().drop("index", axis=1)

shots_df["Shot number"] = shots_df["Shot number"].astype("Int64") #convert shot numbers to 64-bit integer from float
```

```python
# create json file with paths to each of the devices listed in the master csv
# can then edit this json file to add the path to each of the devices

base_directory = "eos/project/h/hiradmat/HRMT Experiments/2024/FIREBALL2/FB2_HRMT64_repository/HRMT64 data"
NON_DEVICE_COL_NAMES = ["Timestamp", "Shot number"]
paths_to_devices = {key: f"{base_directory}/**EDIT_THIS_PART_IN_THE_JSON_FILE**" for key in shots_df.drop(NON_DEVICE_COL_NAMES, axis=1)}


path_to_json = "."
with open(f"{path_to_json}/paths_to_devices.json", "w") as json_file:
    json.dump(paths_to_devices, json_file, indent=4)


# =============================================================================================================


# load the json file back in
with open(f"{path_to_json}/paths_to_devices.json", "r") as json_file:
    paths_to_devices = json.load(json_file)

print(paths_to_devices)
```

```python
# check that it works for every shot in the master csv

# some shot numbers are missing, so have to index using the row rather than the shot number
# finds row with a given shot number. could just use enumerate() but this will be better when searching for a specific shot

for shot in shots_df["Shot number"]:

    row = shots_df[shots_df["Shot number"] == shot].index[0]   # does assume there will never be duplicate shot numbers
    devices_used = shots_df.iloc[row].dropna()
    devices_used = devices_used.drop(NON_DEVICE_COL_NAMES).to_dict()
    complete_filenames = {key : f"{paths_to_devices[key]}/{devices_used[key]}" for key in devices_used}


    print(f"row index: {row}, shot number: {shot}")
    for key, value in complete_filenames.items():
        print(f"{key}: '{value}'")

    print("")
```
