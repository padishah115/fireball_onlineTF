```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = '../example_data/data/1ns_test.dac'
```

```python
def _load_ORCA_image(path:str):

    img = np.genfromtxt(data_path)

    #x axis encodes information about space in mm
    space_mm_x = img[0, 1:]

    #y axis encodes information about time in ns
    time_ns_y = img[1:, 0]
    img = img[1:, 1:]

    return img, space_mm_x, time_ns_y

img, space_mm_x, time_ns_y = _load_ORCA_image(path=data_path)
space_mm_x, time_ns_y
```

```python
fig, axs = plt.subplots()
step = 100

time_ns_y_rounded = [f"{time:.2f}" for time in time_ns_y]

x_ticks_pos = np.arange(0, len(space_mm_x), step=step)
y_ticks_pos = np.arange(0, len(time_ns_y), step=step)
x_ticks = space_mm_x[::step]
y_ticks = time_ns_y_rounded[::step]

axs.imshow(img)
axs.set_xlabel("Distance / mm")
axs.set_ylabel("Time / ns")
axs.set_yticks(y_ticks_pos)
axs.set_yticklabels(y_ticks)
axs.set_xticks(x_ticks_pos)
axs.set_xticklabels(x_ticks)

plt.show()
```
