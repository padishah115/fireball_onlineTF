```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = '../example_data/data/1ns_test.dac'
```

```python
def _load_ORCA_image(path:str):

    img = np.genfromtxt(data_path)

    #x axis encodes information about space in mm
    space_mm_x = img[0, 1:]

    #y axis encodes information about time in ns
    time_ns_y = img[1:, 0]
    img = img[1:, 1:]

    return img, space_mm_x, time_ns_y

img, space_mm_x, time_ns_y = _load_ORCA_image(path=data_path)
```

```python
fig, axs = plt.subplots()

axs.imshow(img)
axs.set_xlabel("Distance / mm")
axs.set_ylabel("Time / ns")

plt.show()
```
